```python
# question 2: retrieve and print all documents from the orders collection.

from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
collection = client["superstoredb"]["orders"]

for doc in collection.find():
    print(doc)
```

```python
# question 4: write a query to fetch all orders from the west region.

from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
collection = client["superstoredb"]["orders"]

for doc in collection.find({"Region": "West"}):
    print(doc)
```

```python
# question 5: write a query to find orders where sales is greater than 500.

from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
collection = client["superstoredb"]["orders"]

for doc in collection.find({"Sales": {"$gt": 500}}):
    print(doc)
```

```python
# question 6: fetch the top 3 orders with the highest profit.

from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
collection = client["superstoredb"]["orders"]

for doc in collection.find().sort("Profit", -1).limit(3):
    print(doc)
```

```python
# question 7: update all orders with ship mode as "first class"
# to "premium class".

from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
collection = client["superstoredb"]["orders"]

result = collection.update_many(
    {"Ship Mode": "First Class"},
    {"$set": {"Ship Mode": "Premium Class"}}
)

print(result.modified_count, "documents updated.")
```

```python
# question 8: delete all orders where sales is less than 50.

from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
collection = client["superstoredb"]["orders"]

result = collection.delete_many(
    {"Sales": {"$lt": 50}}
)

print(result.deleted_count, "documents deleted.")
```

```python
# question 9: use aggregation to group orders by region
# and calculate total sales per region.

from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
collection = client["superstoredb"]["orders"]

pipeline = [
    {
        "$group": {
            "_id": "$Region",
            "Total Sales": {"$sum": "$Sales"}
        }
    }
]

for doc in collection.aggregate(pipeline):
    print(doc)
```

```python
# question 10: fetch all distinct values for ship mode.

from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
collection = client["superstoredb"]["orders"]

print(collection.distinct("Ship Mode"))
```

```python
# question 11: count the number of orders for each category.

from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
collection = client["superstoredb"]["orders"]

pipeline = [
    {
        "$group": {
            "_id": "$Category",
            "Total Orders": {"$sum": 1}
        }
    }
]

for doc in collection.aggregate(pipeline):
    print(doc)
```
